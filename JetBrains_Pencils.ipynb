```python
import random 

def start_game():
  print("How many pencils would you like to use:")
  while True: 
    pen = input()
    if not pen.isnumeric():
      print("The number of pencils should be numeric")
      continue 
    elif int(pen) <= 0:
      print("The number should be positive")
      continue
    pen = int(pen)
    break 
  return pen

def whos_first(players):
  print("Who will be the first(John, Jack):")
  while True:
    player =''
    bot = ''
    n = input()
    if n == players[0]:
      player = players[0]
      bot = players[1]
      break
    if n == players[1]:
      player = players[1]
      bot = players[0]
      break
    else:
      print("Choose between 'John' and 'Jack'")
  return player, bot 

def check_input(value): 
  valid = True
  if not value.isnumeric():
    print("The number of pencils should be numeric")
    print("Possible values: '1', '2', '3'")
    valid = False
  elif int(value) <= 0:
    print('The number of pencils should be positive')
    print("Possible values: '1', '2', '3'")
    valid = False
  elif int(value) > 3:
    print("Possible values: '1', '2', '3'")
    valid = False
  return valid

students = ['John', 'Jack'] # list with players names
pen = start_game() # start game, input how many pencils
player, bot = whos_first(students)  # check who is first and return names  
print(pen * '|')
count = 0
while pen > 0:
  if count % 2 == 0:
    print(player +"'s turn!")
    took = input()
    if not check_input(took):  # check valid input. if not valid -> next iteration
      continue
    else:
      pen -= int(took)
      print('|' * pen)
  else:
    print(bot + "'s turn")
    bot_took = [1, 2, 3]
    if pen % 4 == 0:
      bot_took = 3
    if pen % 4 == 3:
      bot_took = 2
    if pen % 4 == 2:
      bot_took = 1 
    pen -= bot_took
    print(bot_took)
    print('|' * pen)
  
  if pen == 0:
    print(bot +" won!" if count % 2 == 1 else  player + " won!")
  
  count += 1 
```
